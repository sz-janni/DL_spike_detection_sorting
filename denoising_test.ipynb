```python
import pickle

with open('./spike_waveform_data.pickle', 'rb') as f:
        pure = pickle.load(f)
        pure = pure[::100]

dn.train_model(pure, noise_factor=0.4)
```

```python
noisy = dn.add_noise(pure, 0.4)

cleared = dn.denoise_signal( noisy )


from matplotlib import pyplot as plt
import numpy as np


preview_cnt = 3
fig, axes = plt.subplots(preview_cnt, 3)
for i in range(preview_cnt):
    axes[i, 0].plot(noisy[i])
    axes[i, 0].set_title('Noisy')

    axes[i, 1].plot(pure[i])
    axes[i, 1].set_title('Pure')

    axes[i, 2].plot(cleared[i])
    axes[i, 2].set_title('Denoised')
plt.tight_layout()
plt.show()
```
